```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv', index_col = 0)
```

```python
categoricals = df.select_dtypes(object)
categoricals.head()
```

```python
for col in categoricals.columns:
    categoricals[col].value_counts().plot.bar(rot=0)

    plt.show()
```
